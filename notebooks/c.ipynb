```python
runner = Runner(
    Input,
    Result,
    solver_cmd="solver_cmd",
    solver_version="solver_version",
    database_csv="../log/database.csv",
    input_csv="../log/input.csv",
    log_file="../log/a.log",
    is_maximize=False,
    verbose=0,
)
df = runner.read_database_df()
df = df[df.solver_version.str.startswith("multi-start-")]
df.solver_version = df.solver_version.str[len("multi-start-"):].astype(int)
df
```

```python
for i in range(1, 11):
    a = df[df["n"] // 5 == i]
    if len(a) == 0:
        continue
    print(i, a.groupby("solver_version").relative_score.mean().idxmax())
```

```python
for i in range(1, 11):
    a = df[df["d"] // 5 == i]
    if len(a) == 0:
        continue
    print(i, a.groupby("solver_version").relative_score.mean().idxmax())
```

```python
for i in range(75, 100):
    a = df[(df["a.sum.mean"] / 10000).astype(int) == i]
    if len(a) == 0:
        continue
    print(i, a.groupby("solver_version").relative_score.mean().idxmax())
```
